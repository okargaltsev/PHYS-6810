```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

from hr_evolution_widget_mistgrid_absolute_widgetfixed import (
    build_population,
    draw_snapshot,
)

pop = build_population(mean_feh=0.0, feh_scatter=0.10)

age_slider = widgets.FloatSlider(
    value=140.0,
    min=1.0,
    max=3000.0,
    step=2.0,              # <-- fixed step in Myr; change this
    description="Age [Myr]",
    continuous_update=False,
    readout_format=".0f",
)

play = widgets.Play(
    value=0,
    min=0,
    max=int((age_slider.max - age_slider.min) / age_slider.step),
    step=1,
    interval=120,           # <-- milliseconds between frames; change this
    description="Play",
)

def _sync_play_to_slider(change):
    age_slider.value = age_slider.min + change["new"] * age_slider.step

play.observe(_sync_play_to_slider, names="value")

out = widgets.Output()

def redraw(age_myr):
    with out:
        out.clear_output(wait=True)
        plt.close("all")
        fig, ax = draw_snapshot(pop, age_myr=age_myr)
        display(fig)
        plt.close(fig)

widgets.interactive_output(
    redraw,
    {"age_myr": age_slider}
)

display(widgets.VBox([
    widgets.HBox([play, age_slider]),
    out
]))

redraw(age_slider.value)
```
