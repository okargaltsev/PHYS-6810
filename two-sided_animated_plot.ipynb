```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

x = np.linspace(0, 4*np.pi, 1000)

def f(a):
    fig, ax1 = plt.subplots(figsize=(8,4))
    ax2 = ax1.twinx()

    ax1.plot(x, np.sin(x+a), 'C0', lw=2)
    ax2.plot(x, 3*np.sin(x) + 5*np.cos(x), 'C1', lw=2)

    ax1.set_xlim(0, 4*np.pi)
    ax1.set_ylim(-1.2, 1.2)
    ax2.set_ylim(-6.5, 6.5)
    ax1.grid(True, alpha=0.3)
    plt.show()

interact(f, a=FloatSlider(min=0, max=2*np.pi, step=0.01, value=0));
```

### The above is not ideal because axis are of the same color. 

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

x = np.linspace(0, 4*np.pi, 1000)

def plot_demo(a):
    fig, ax1 = plt.subplots(figsize=(9, 4.5))
    ax2 = ax1.twinx()

    line1, = ax1.plot(x, np.sin(x + a), color='C0', lw=2, label=r'$\sin(x+a)$')
    line2, = ax2.plot(
        x, 3*np.sin(x) + 5*np.cos(x),
        color='C1', lw=2, label=r'$3\sin x + 5\cos x$'
    )

    ax1.set_xlim(0, 4*np.pi)
    ax1.set_ylim(-1.2, 1.2)
    ax2.set_ylim(-6.5, 6.5)

    ax1.set_xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi])
    ax1.set_xticklabels(['0', r'$\pi$', r'$2\pi$', r'$3\pi$', r'$4\pi$'])

    ax1.set_xlabel('x')
    ax1.set_ylabel(r'$\sin(x+a)$', color='C0')
    ax2.set_ylabel(r'$3\sin x + 5\cos x$', color='C1')

    ax1.tick_params(axis='y', colors='C0')
    ax2.tick_params(axis='y', colors='C1')
    ax1.spines['left'].set_color('C0')
    ax2.spines['right'].set_color('C1')

    ax1.grid(alpha=0.3)
    ax1.set_title(r'Interactive check: $3\sin(x)+5\cos(x)=A\sin(x+\phi)$')

    lines = [line1, line2]
    labels = [line.get_label() for line in lines]
    ax1.legend(lines, labels, loc='upper right')

    plt.show()

interact(
    plot_demo,
    a=FloatSlider(value=0, min=0, max=2*np.pi, step=0.01, description='a')
)
```
