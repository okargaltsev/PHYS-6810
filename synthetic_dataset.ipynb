```python
import numpy as np
from numpy.random import rand, randn
import matplotlib.pyplot as plt
```

```python
# Generate an artificial data set shown during the lecture
```

```python
ng = 50

g1 = np.transpose([randn(ng) * 0.1 - 3, randn(ng) + 1])
g2 = np.transpose([randn(ng) * 0.1 + 3, randn(ng) + 1])

gtemp = np.transpose([randn(2 * ng) * 0.75, randn(2 * ng) * 0.75])
g3, g4 = np.split(gtemp[np.argsort(np.sum(gtemp**2, axis=1))], 2)

g51 = np.linspace(-2.75, 2.75, ng)
g52 = (g51 / 2)**2 - 4
g5 = np.transpose([g51, g52 + randn(ng) * 0.5])

# We got a 5 subsets of data:
data1 = np.array([g1, g2, g3, g4, g5])
# Join all subsets into one:
data = data1.reshape(-1, data1.shape[-1])
```

```python
plt.scatter(*data.T, c='k')
plt.show()
```
